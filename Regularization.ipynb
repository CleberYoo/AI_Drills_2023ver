# Regularization
## Ridge : 계수를 제곱한 값을 기준으로 규제
### (계수 > 1) -->> 극대화 / (계수 < 1) -->> 극소화
## Lasso : 계수의 절댓값을 기준으로 규제 -->> 순해짐

# Ridge

```python
from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.fit(train_input_scaled, train_target)

print(ridge.score(train_input_scaled, train_target))
print(ridge.score(test_input_scaled, test_target))
```

```python
import matplotlib.pyplot as plt

train_score = []
test_score = []

alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]

for alpha in alpha_list:
    ridge = Ridge(alpha= alpha)   ### 릿지 모델 객체 생성
    ridge.fit(train_input_scaled, train_target) ### 릿지 모델 훈련
    train_score.append(ridge.score(train_input_scaled, train_target))   ### 훈련 점수와 테스트 점수를 저장
    test_score.append(ridge.score(test_input_scaled, test_target))
```

```python
plt.plot(np.log10(alpha_list), train_score)   ### 파란색 선
plt.plot(np.log10(alpha_list), test_score)   ### 주황색 선
plt.xlabel("ALPHA")
plt.ylabel("R^2")
plt.show()
```

### np.log : 밑이 자연상수 e
### np.log10 : 밑이 10

#### alpha 값 중 사용하기 적절한 alpha = 0.1이므로 0.1로 최종 모델을 훈련해보겠다

```python
ridge = Ridge(alpha= alpha_list[2])   ### alpha_list[2] = 0.1

ridge.fit(train_input_scaled, train_target)

print(ridge.score(train_input_scaled, train_target))
print(ridge.score(test_input_scaled, test_target))
```

```python
plt.plot(alpha_list, train_score)   ### 파란색 선
plt.plot(alpha_list, test_score)   ### 주황색 선

plt.xlabel("ALPHA")
plt.ylabel("R^^2")

plt.show()
```

# Lasso

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(train_input_scaled, train_target)

print(lasso.score(train_input_scaled, train_target))
print(lasso.score(test_input_scaled, test_target))
```

```python
train_score2 = []
test_score2 = []

alpha_list2 = [0.001, 0.01, 0.1, 1, 10, 100]

for alpha in alpha_list2:
    lasso = Lasso(alpha= alpha, max_iter=10000)
    lasso.fit(train_input_scaled, train_target)
    train_score2.append(lasso.score(train_input_scaled, train_target))
    test_score2.append(lasso.score(test_input_scaled, test_target))
```

```python
plt.plot(np.log10(alpha_list2), train_score2)
plt.plot(np.log10(alpha_list2), test_score2)

plt.xlabel("ALPHA")
plt.ylabel("R^2")
plt.show()
```
