```python
plt.imshow(fruits[0], cmap='gray_r')   # 흑백이미지로 0번째 과일 이미지출력
plt.show()
```

```python
#픽셀 값 100X100 -> 10,000으로 재배열하기(=차원 축소/확대) reshape

apple = fruits[0:100].reshape(100, 100*100)   #3차원(300, 100, 100) -> 2차원(300, 10000) 차원축소
pineapple = fruits[100:200].reshape(-1, 100*100)   #1번째 차원을 100에서 -1로 바꿔주었는데 둘이 동일하다. -1의 의미는 남은 차원을 할당한다는 뜻
banana = fruits[200:300].reshape(-1, 100*100)
```

```python
print(apple.shape, pineapple.shape, banana.shape)
```

```python
print(apple.mean(axis=1))   #apple의 각 행(0~100개 사과)의 열(픽셀 값)들을 평균
print(pineapple.mean(axis=1))   #pineapple의 각 행(0~100개 사과)의 열(픽셀 값)들을 평균
print(banana.mean(axis=1))   #banana의 각 행(0~100개 사과)의 열(픽셀 값)들을 평균
```

```python
#히스토그램 막대그래프로 각 행별 열(픽셀 값)들의 평균 시각화 해보기

plt.hist(np.mean(apple, axis=1), alpha = 0.75, color='#0044FF')   #alpha = 투명도
plt.hist(np.mean(pineapple, axis=1), alpha = 0.75, color='#FF4400')
plt.hist(np.mean(banana, axis=1), alpha = 0.75, color='#44FF00')
plt.ylim(0, 30)
plt.xlabel('PIXEL')
plt.ylabel('AMOUNT')
plt.show()
```

```python
#바 그래프로 각 열(픽셀값) 별 행(각 과일별 100개)들의 평균 시각화 해보기

fig, axs = plt.subplots(1, 3, figsize=(25,5))
axs[0].bar(range(0, 10000), np.mean(apple, axis=0))
axs[1].bar(range(0, 10000), np.mean(pineapple, axis=0))
axs[2].bar(range(0, 10000), np.mean(banana, axis=0))
plt.show()
```

```python
apple_mean = np.mean(apple, axis=0).reshape(100, 100)
pineapple_mean = np.mean(pineapple, axis=0).reshape(100, 100)
banana_mean = np.mean(banana, axis=0).reshape(100, 100)

fig, axs = plt.subplots(1, 3, figsize=(20,20))

axs[0].imshow(apple_mean, cmap='gray')
axs[1].imshow(pineapple_mean, cmap='gray')
axs[2].imshow(banana_mean, cmap='gray')
plt.show()
```

```python
abs_diff = np.abs(fruits - apple_mean)
abs_mean = np.mean(abs_diff, axis=(1,2))   # inedx 0차원(300개) 1차원(100개) 2차원(100개) 중 index 1,2차원을 가지고 오차 평균을 구함   # 각 샘플의 오차 평균이므로 abs_mean = (300, ) 1차원 배열
```

```python
abs_mean.shape
```

```python
abs_diff.shape
```

```python
apple_mean.shape
```

```python
apple_index = np.argsort(abs_mean)[:100]   # 총 300개 과일을 사과 평균값을 빼고 절댓값을 한 abs_mean을 정렬하고 0~99개를 apple_index에 저장
                                           # 정렬은 오름차순이므로 절댓값 오차가 제일 적은 애가 배열 앞쪽에 위치하게 된다.
                                           # 그러면 사과들이 사과 평균값과 뺏을 때 0에 수렴할테니 배열 앞쪽에 위치하여 apple_index에 사과들이 저장된다.
fig, axs = plt.subplots(10, 10, figsize = (25, 25))   #10X10 배열 생성
for i in range(10):   # i = 행
  for j in range(10):   # j = 열
    axs[i, j].imshow(fruits[apple_index[i * 10 + j]], cmap='gray')

plt.show()
```
